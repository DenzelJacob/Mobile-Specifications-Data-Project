```python
import pandas as pd
import numpy as np
```

```python
adf = pd.read_csv("mobiles1.csv", usecols =['name', 'price', 'rating','memory']) # android raw
idf = pd.read_csv("apple_products.csv", usecols=['Product Name','Mrp','Star Rating','Ram']) # iphone raw

adf['Ram'] = adf['memory'].str[:1]
adf = adf.drop(columns=['memory'])

idf['Ram'] = adf['Ram'].str[:1]

#   1 USD = 75.31 INR
adf['Price_USD']=adf.price/75.31
adf['Rating per dollar'] = adf.rating / adf.Price_USD

idf['Price_USD'] = idf['Mrp']/75.31
idf['Rating per dollar'] = idf['Star Rating'] /idf['Price_USD']

adf['Ram per Dollar'] = adf['Ram'].astype(int) /adf['Price_USD']
idf['Ram per Dollar'] = idf['Ram'].astype(int) /idf['Price_USD']

idf=idf.rename(columns={"Product Name": "name", "Mrp": "price",'Star Rating':'rating'})
pass
```

```python
df=adf.append(idf).reset_index()
df
```

^^^^ combine both datasets

The two criterion I selected were phones with ram amount greater than 4 GB, I did this to select the phones that I would consider powerful enough for most task.
The second criterion I selected were phones that cost above $500,  this way the quality of the phones were guaranteed to be decent.

```python
df1= df[df['Ram'].astype(int) > 4][['name','Ram']]
df1
# s1 is Ram > 4gb
```

```python
df2= df[df['Price_USD'].astype(float) > 500][['name','Price_USD']]
df2
#s2 is Price_USD > 500
```

```python
df_intersection = pd.merge(df1,df2, on='name')[['name','Ram','Price_USD']]
df_intersection
#default inner join is enough for intersection
```

^^^ in order to have an intesection of both dataframes I used the merge on the name column,
this created a smaller dataset that the two original ones as it on would count the phones with both, more than 4GB Ram and a price greater than $500


```python
pd.merge(df1,df2, how ='outer')
```

^^^ If i were to just combine the datasets with an outer join I would lose the opposite data for each dataset

```python
df_union= pd.merge(df[df['Ram'].astype(int) > 4],df[df['Price_USD'].astype(float) > 500],on = 'name', how ='outer')
df_union['Ram']=df_union['Ram_x'].combine_first(df_union['Ram_y'])
df_union['Price_USD']=df_union['Price_USD_x'].combine_first(df_union['Price_USD_y'])
df_union = df_union[['name','Ram','Price_USD']]
# union by outer join with replacement
df_union
```

^^^ In order to fill the missing spots I merged the two subsets of the main dataset with the same criterion as df1 and df2, but these subsets contain the information of both S1 and S2 at the same time. I then used the .combine_first() method to fill in the missing spots, and then I presented the new dataset in a proper format.
